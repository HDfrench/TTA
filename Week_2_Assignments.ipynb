```python
### Week 2 - python programming assignment

# this function returns 2 values:
    # the converted input string as an integer (or a float if typ is declared as "f"
    # and a boolean value
def check_valid_input(inp, typ='i'):
    try:
        # tries to convert the input into an integer if typ is "i"
        if typ == "i":
            convert = int(inp)
        # or tries to convert fo float if typ is "f"
        elif typ == "f":
            convert = float(inp)
        else:
        # if another undeclared typ variable is passed
        # raise an error
            raise ValueError
        # if successful, the Boolean value is set to True and the integer is returned (convert)
        return True, convert
    except:
        # if an error occurred in the conversion process
        # the boolean value is set to False
        # and convert to None
        return False, None
```

```python
# this function is mare to check if an integer is within an appropriate range:
    #it has 4 parameters (1 mandatory and 3 optional:
        # the value to be checked
        # a string containing a message in case the value is outside the range
        # the lower limit, define by default at 0
        # the upper limit, define by default at 1000000
    # return a boolean value True if in appropriate range, else False
def check_my_range(val, stg = '',  low_limit = 0, high_limit = 1000000):
    if val in range(low_limit, high_limit + 1):
        return True
    else:
        print(stg)
        return False
```

```python
# procedure to calculate the yearly depreciation of goods
# the argument is a list
def show_depreciation(lst):
    # read some items from the list passed as an argument and assign them to so local variables
    item, year, price, target = lst[0], lst[1], lst[2], lst[3]
    # print the initial statement (header)
    # but checks whether the item name starts with a vowel
    if item[0] in ["aeiou"]:
        print(f"If an {item} costs £{price} in {year}, it's value will be:")
    else:
        print(f"If a {item} costs £{price} in {year}, it's value will be:")
    while price > target:
        price *=0.9
        year +=1
        print(f"{year}\t £{price:.2f}")
```

```python
# this procedure is the mini calculator that requires a list as parameter (list of 2 numbers provided by the user)
def mini_calculator(numbers):
    # dictionary defined as a reference
    # it could have been defined outside the procedure and declared as global variable here
    tasks = {"+": ["+","sum", "plus"],
            "-": ["-", "substraction", "minus"],
            "*": ["*", "product", "times"],
            "/": ["/", "division", "by"],
            "^": ["^", "factorisation", "to the power of"]}
    # string used for prompting the user
    stg ="""Select an operator:
        \t+ for addtion
        \t- for substraction
        \t* for multiplication
        \t/ for division
        \t^ for factorisation\n"""
    # this is an embedded fuction to do the required calculation
    # it requires the list of numbers the user provided and the selected operator
    # this returns "result" which is the resulting calculation
    def calculate(operator, numbers):
        # we now check which 1st item in our selected list corresponds to the user choice
        # (it might seems strange to have a duplication of the key as the first item in the list, but python makes it difficult to lookup keys)
        # the the relevant operator is located, a result is calculated based on the user choices of numbers and operator
        if operator == "+":
            result = numbers[0] + numbers[1]
        elif operator == "-":
            result = numbers[0] - numbers[1]
        elif operator == "*":
            result = numbers[0] * numbers[1]
        # if the operator is a division we include an additional check
        elif operator == "/":
            # to avoid a division by 0 error
            if numbers[1] != 0:
                result = numbers[0] / numbers[1]
            else:
                # if the second digit is 0, result returns None\n",
                result = None
                # and we print a message to that effect to the user
                print("Division by 0!\nImpossible!!")
        else:
            result = numbers[0]**numbers[1]        
        return result
    
    # This is a generic while loop that will be always True
    # You have to be careful with this as if you omit to break it, you can end up with a perpetual loop
    while True:
        # we ask the user which operand is to be used
        oper = input(stg)
        # we check if that selection is one of of our dictionary keys
        if oper in tasks.keys():
            # if that is the case, we escape the while loop
            break
        # otherwise the loop continues until the user submits a valid choice

    # now that we have collected the values and the operator from the user, we can calculate and display the result
    mycalculation = tasks[oper]
    # we check that the operation did not return a None value (in which case the internal function would have printed a message)
    if calculate(mycalculation[0], numbers) !=None:
        # and print the result using again positional variables
        print("The {0} of {1} {2} {3} is {4}".format(mycalculation[1], numbers[0], mycalculation[2], numbers[1], calculate(mycalculation[0], numbers)))
```

```python
#generic user input function
def user_input(selection):
    # if selection is 1 or 2,
    # we need a single user input
    if selection in [1, 3, 5]:
        #loop to get user input and ensure it is a valid number
        while True:
            #request user input
            number = input("Enter a number: ")
            valid, numb = check_valid_input(number)
            # if valid returns True the loop breaks
            #otherwise continues
            if valid == True:
                break
        # based on the value of the selection, we call the relevant function
        if selection ==1:
            # call the procedure number_checker using the user input as an argument
            number_checker(numb)
        elif selection == 3:
            # call the number_tallier function using
            # the user input, 1 (to increment), the maximum number and a string (greater)
            number_tallier(numb, 1, 10, "greater")
        elif selection == 5:
            # call the number_tallier function using
            # the user input, -1 (to decrement), the maximum number and a string (lower)
            number_tallier(numb, -1, 0, "lower")
        return True
    # check if the user input is 2 or 4 to display the pseudo code
    elif selection in [2,4]:
        # call the procedure to print the pseudo code
        if selection == 2:
            # the variable 0 means to read the first list in the specific list
            conditional_pseudo_code(0)
        elif selection  == 4:
            # the variable 1 means to read the second and last list in the specific list
            conditional_pseudo_code(1)
        return True
    elif selection == 6:
        items = []
        while len(items)< 4:
            if len(items) == 0:
                inp = input("What object are you looking at assessing? ")
                if len(inp) > 0:
                    items.append(inp)
            elif len(items) == 1:
                inp = input("What year was is purchased? ")
                valid, year = check_valid_input(inp)
                if valid == True:
                    if check_my_range(year,"The years range from 1960 to 2020", 1960, 2021) == True:
                        items.append(year)
            elif len(items)==2:
                inp = input("What was the original purchase price in pounds? ")
                valid, price = check_valid_input(inp)
                if valid == True:
                    if check_my_range(price, "The minimum price should be £500", 500) == True:
                        items.append(price)
            elif len(items) == 3:
                inp = input("What would be the minimum depreciation value? ")
                valid, depre = check_valid_input(inp)
                if valid == True:
                    if check_my_range(depre, "The minimum value for the calculation should be £250", 250) == True:
                        if depre < price:
                            items.append(depre)
                        else:
                            print("Sorry, but the depreciated value cannot be higher than the purchase price")
        # we call the function to display the depreciation 
        show_depreciation(items)
        return True
    elif selection == 7:
        # we define an empty list to collect the user input
        numbers = []
        # this while loop will continue until the user has entered 2 valid numbers (in our case, we accept integer AND floats)
        while len(numbers)<2:
            # if our reference list is empty (numbers)
            if len(numbers) == 0:
                # we ask the user for the first number
                nb = input("Input the first number: ")
            else:
                # otherwise we ask them for the second number
                nb = input("Input the second number: ")
            # after each user input, we check the validity in the try section
            valid, val = check_valid_input(nb, "f")
            if valid == True:
                numbers.append(val)
        # we call our mini calculator function
        mini_calculator(numbers)
        return True
    # if the user selection is not amongst the previous options (1 to 7)
    # returns False
    else:
        return False
```

```python
# procedure to check if a number is > 10 or not
def number_checker(nb):
    #This is a simple procedural definition that prints whether a number
    #given as a parameter is higher or lower than 10
    if nb > 10:
        print("Number is greater than 10")
    else:
        print("Number is smaller than 10")
```

```python
# procedure to increment or decrement a number until a certain target is reached
def number_tallier(number, increment, target, stg):
    #set a boolean value to False
    satisfied = False
    # this is another procedural routine
    # which checks if a number is not equal to a target using 2 logical expressions
    if (increment == -1 and number < target) or (increment == 1 and number > target):
        print("Your number ('{}') is {} than {}".format(number, stg, target))
        satisfied = True
    if number == target:
        print("Your number ('{}') is equal to {}".format(number, target))
        satisfied = True
    if satisfied == False:
        # if not, it prints a statement
        print("The {} is not yet {} than {}, but soon!".format(number, stg, target))
        # and call itself increasing/decreasing the number
        number_tallier(number + increment, increment, target, stg)
```

```python
# procedure to display the pseudo code for selection 2 and 4
def conditional_pseudo_code(ref):
    # this list contains strings that belong to both pseudo codes
    core_start = ["start", "Step1:" + " " * 5 + "Ask user to input an integer",
            "Step 2:" + " " * 5 + "check it is a valid integer",
            " " * 12 + "No?"  + " " * 5 + "ask for another user input",
            " " * 20 + "repeat Step 1"]
    # this list contains 2 lists, each for a specific pseudo code
    # the variable ref passed in calling this function defines which
    # internal list will be read (index 0 or index 1)
    lines = [["Step 3:" + " " * 5 + "check if the number is greater than 10",
            " " * 12 + "Yes?"  + " " * 4 + "print 'number greater than 10'",
            " " * 12 + "No?"  + " " * 5 + "print 'number lower than 10'",
            " " * 20 + "substract 1 to the number"],
            ["Step 3:" + " " * 5 + "check if the number is greater than 0",
            " " * 12 + "Yes?"  + " " * 4 + "print 'number greater than 0'",
            " " * 12 + "No?"  + " " * 5 + "print the number",
            " " * 20 + "add 1 to the number"]]
    core_end = [" " * 12 + "Yes?" + " " * 4 + "execute step 3",
            " " * 20 + "repeat Step 3", "end"]
    # small internal procedure to loop through the lists of text
    def read_core(lst):
        for line in lst:
            print(line)
    # read and print the first lines of the common list (all the items except the last 2
    read_core(core_start)
    # read and print all the line specific to the requested pseudo code
    # based on the variable passed on during user selection
    read_core(lines[ref])
    # read the last section of the pseudo code
    read_core(core_end)
```
